## Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for Python. It is designed to be lightweight and easy to use, making it an excellent choice for developing web applications and APIs quickly. Flask does not include all the components and features that you might find in a full-stack web framework like Django, but it provides the core functionality necessary for building web applications. Some of the key advantages of Flask include:

- Lightweight: Flask is a micro-framework, which means it comes with minimal built-in features and libraries. This gives developers the flexibility to choose and integrate the specific components they need, allowing for a more customized and lightweight application.

- Simplicity: Flask has a simple and straightforward API, making it easy for developers to learn and use. This simplicity is one of its most attractive features, especially for those new to web development.

- Extensibility: Flask is designed to be extensible. You can add various third-party extensions and libraries to enhance its functionality. This makes it easy to integrate features like database support, authentication, and more, as needed for your project.

- Flexibility: Flask doesn't impose a strict project structure or folder hierarchy, giving developers the freedom to organize their code as they see fit. This flexibility can be an advantage for smaller projects or when you have a specific project structure in mind.

## Q3. What is App routing in Flask? Why do we use app routes?


In Flask, app routing refers to the process of defining URL routes for your web application. A route is a mapping between a URL and a Python function. When a user visits a specific URL in your web application, Flask's routing mechanism determines which Python function should handle the request.

Here's how app routing works in Flask:

- URL Rules: In Flask, you define URL rules using the @app.route decorator. These rules specify the URLs that your application will respond to and the associated Python functions (view functions) that will be executed when a user visits those URLs. The route decorator is used to bind a function to a specific URL pattern.

- Handling Requests: When a user makes an HTTP request to your application, the Flask framework matches the URL of the request to the defined URL rules and calls the associated Python function (view function). This function handles the request, processes data, and generates a response.

- Dynamic Routes: Flask allows you to define dynamic routes by including placeholders in the URL patterns. For example, you can define routes like /user/<username> where <username> is a placeholder that captures the value from the URL and passes it as an argument to the view function. This enables you to create routes that handle a variety of inputs.
    
App routing is essential in Flask for several reasons:

1. Organizing URLs: It allows you to define a clear and organized structure for your web application, making it easy to navigate and understand.

2. Separation of Concerns: By mapping specific URLs to dedicated view functions, you can separate the logic of your application, making it more maintainable and modular.

3. Parameter Handling: You can capture parameters from the URL and use them in your view functions, allowing your application to respond to dynamic data and user input.

4. RESTful API Development: App routing is fundamental when developing RESTful APIs, where different routes correspond to various resources and actions.

5. User Experience: Routing is crucial for creating a user-friendly and intuitive web application with meaningful URLs.
